##Formulas

```python
# M sera el Resultado de la comparación
# Calculando que valor es mayor
# Datos de entrada A , B
# Mostrar el numero mayor
# Si es verdadero mostrar True
A = int(input("Ingrese el primer valor: "))
B = int(input("Ingrese el segundo valor: "))
M = A > B
if M:
    print("El primer valor es mayor")
else:
    print("El segundo valor es mayor")

print(f"El resultado de la comparación (M) es: {M}")
print("Fin del programa")
print("Programa realizado por Jaquez Andres NL = 27")
```
